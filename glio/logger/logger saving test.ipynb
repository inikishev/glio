```python
from glio.logging.logger import Logger
import torch, numpy as np
import random
```

```python
logger = Logger()
for i in range(1000): logger.add("metric 1", random.random(), i)
logger.plot("metric 1")
```

```python
for i in range(100): logger.add("metric 2", random.random(), i*10)
logger.plot("metric 2")
```

```python
logger.save("test save.npz")
```

```python
restored = Logger.from_file("test save.npz")
```

```python
logger.plot("metric 1", "metric 2")
```

```python
restored.plot("metric 1", "metric 2")
```
