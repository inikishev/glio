```python
from glio.optim.opteval import OTImageReconstruct
from glio.optim.randomsearch import RandomSearch, RandomStepSearch, RandomDeltaSearch, HillClimbing, RandomDeltaLossSearch, RandomMomentumSearch, ClosureOptimizer, ConvFilterSearch, clone_params,set_params_
import matplotlib.pyplot as plt
from glio.jupyter_tools import show_slices
from glio.visualize import vis_imshow
from glio.plot import *
from glio.random import *
from glio.loaders.image import read as imread
import torch, numpy as np, random
from torchvision.transforms import v2
```

```python
image = v2.Resize((32,32), antialias=False)(imread(r"C:\Users\nkshv\Pictures\AA_AerofleetArt.jpg"))
vis_imshow(image)
```

```python
test = OTImageReconstruct(image, init=torch.zeros_like, loss=torch.nn.MSELoss())
optimizer = RandomSearch(test.parameters(), rand=torch.randn_like)
print("testing")
with torch.no_grad():
    test(optimizer, 50000)
print("Done")
qlinechart(test.losses)
plt.show()
show_slices(test.preds)
qpath2d(test.preds)
```

```python
test = OTImageReconstruct(image, init=torch.zeros_like, loss=torch.nn.MSELoss())
optimizer = RandomStepSearch(test.parameters(), lr=4e-3, rand=torch.randn_like)
print("testing")
with torch.no_grad():
    test(optimizer, 50000)
print("Done")
qlinechart(test.losses)
plt.show()
show_slices(test.preds)
qpath2d(test.preds, det=True)
```
